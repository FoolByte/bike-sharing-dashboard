## Pola Penyewaan Sepeda Berdasarkan Waktu

Jam Sibuk:
* Penyewaan tertinggi terjadi pada pagi hari (07:00 - 09:00) dan sore hari (17:00 - 19:00), menunjukkan bahwa banyak pengguna menggunakan sepeda untuk pergi dan pulang kerja/sekolah.
* Pada malam hari (00:00 - 05:00), jumlah penyewaan sangat rendah .

Bulan dan ️ Musim:
* Penyewaan sepeda meningkat mulai Maret hingga Oktober, dengan puncak di musim panas.
* Penyewaan paling rendah terjadi di musim dingin (Desember - Februari), kemungkinan besar karena cuaca yang tidak mendukung.
* Musim gugur (Fall) memiliki jumlah penyewaan tertinggi, mungkin karena suhu yang nyaman untuk bersepeda .

## Pengaruh Cuaca terhadap Penyewaan

Cuaca Cerah vs. ️ Hujan/ ️ Salju:
* Cuaca cerah meningkatkan jumlah penyewaan secara signifikan.
* Cuaca mendung masih cukup tinggi, meskipun sedikit lebih rendah dari cuaca cerah ️.
* Hujan dan salju menyebabkan penurunan drastis dalam jumlah penyewaan, menunjukkan bahwa pengguna cenderung menghindari bersepeda dalam kondisi cuaca buruk.

## Pengaruh Faktor Lingkungan (Suhu, Kelembaban, dan Angin)

Suhu (temp):
* Memiliki korelasi positif dengan penyewaan sepeda, artinya semakin hangat suhu, semakin banyak orang yang menyewa sepeda .

Kelembaban (hum):
* Tidak ada pola linier yang jelas. Titik-titik data tersebar tanpa membentuk pola garis lurus yang jelas. Ini menunjukkan bahwa kelembaban mungkin bukan faktor tunggal yang menentukan jumlah penyewaan sepeda.

Kecepatan angin (windspeed):
* Tidak terlalu berpengaruh, menunjukkan bahwa angin bukan faktor utama dalam keputusan penyewaan ️.

## Visualization & Explanatory Analysis

### Pertanyaan 1: Kapan waktu tersibuk dalam penyewaan sepeda?

```python
# Konversi kolom 'hr' ke kategori waktu
hour_df['hr'] = pd.Categorical(hour_df['hr'], ordered=True)
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=hour_df['hr'], y=hour_df['cnt'], color='gray')
sns.lineplot(x=hour_df['hr'], y=hour_df.groupby('hr', observed=False)['cnt'].mean(), color='blue', marker='o', linestyle='dashed')
plt.title("Waktu Tersibuk dalam Penyewaan Sepeda")
plt.xlabel("Jam dalam Sehari")
plt.ylabel("Jumlah Penyewaan")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
colors = ["#8FBC8F" if i in [3, 4, 5] else  # Spring
          "#FFD700" if i in [6, 7, 8] else  # Summer
          "#A52A2A" if i in [9, 10, 11] else  # Fall
          "royalblue" for i in range(12)]  # Winter

sns.barplot(x="mnth", y="cnt",hue="mnth", data=hour_df, estimator=np.mean, errorbar=None, palette=colors)
plt.title("Hubungan Bulan dan Musim Dengan Penyewaan", fontsize=14)
plt.xlabel("Bulan", fontsize=12)
plt.ylabel("Rata-rata Penyewaan", fontsize=12)
plt.xticks(range(12), labels=["Jan", "Feb", "Mar", "Apr", "Mei", "Jun", "Jul", "Agu", "Sep", "Okt", "Nov", "Des"])
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Tambahkan keterangan warna musim
import matplotlib.patches as mpatches
legend_labels = [mpatches.Patch(color="#8FBC8F", label="Spring"),
                 mpatches.Patch(color="#FFD700", label="Summer"),
                 mpatches.Patch(color="#A52A2A", label="Fall"),
                 mpatches.Patch(color="royalblue", label="Winter")]
plt.legend(handles=legend_labels, title="Musim", loc="upper left")

plt.show()
```

### Pertanyaan 2:  Bagaimana pengaruh cuaca dan suhu lingkungan terhadap jumlah penyewaan?

```python
plt.figure(figsize=(8, 5))
weather_labels = ["Cerah", "Mendung", "Hujan", "Badai"]
weather_colors = ["gray", "gray", "gray", "gray"]

avg_rent_per_weather = hour_df.groupby("weathersit", observed=False)["cnt"].mean()

weather_data = pd.DataFrame({
    "weather_labels": weather_labels,
    "avg_rent": avg_rent_per_weather.values
})

sns.barplot(x="weather_labels", y="avg_rent", data=weather_data, palette=weather_colors, hue="weather_labels", legend=False) # hue dan data ditambahkan

plt.title("Jumlah Penyewaan Sepeda Berdasarkan Kondisi Cuaca", fontsize=14)
plt.xlabel("Kondisi Cuaca", fontsize=12)
plt.ylabel("Rata-rata Penyewaan", fontsize=12)
plt.ylim(0, max(avg_rent_per_weather) * 1.2)


plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.show()
```

```python
hour_df["temp_category"] = pd.cut(hour_df["temp"], bins=5, labels=["Sangat Dingin", "Dingin", "Sejuk", "Hangat", "Panas"])
avg_rent_by_temp = hour_df.groupby("temp_category", observed=True)["cnt"].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(x="temp_category", y="cnt", data=avg_rent_by_temp, marker='o', color='gray', linewidth=2.5, markersize=8)

plt.title("Pengaruh Suhu terhadap Jumlah Penyewaan Sepeda", fontsize=16)
plt.xlabel("Kategori Suhu", fontsize=14)
plt.ylabel("Rata-rata Penyewaan", fontsize=14)

plt.grid(True, linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

# INSIGHT DARI ANALISIS DATA PENYEWAAN SEPEDA

Berdasarkan eksplorasi data dan visualisasi yang telah kita lakukan, berikut adalah beberapa temuan utama:

## 1. Pengaruh Musim terhadap Penyewaan Sepeda

* Penyewaan sepeda tertinggi terjadi di musim gugur (fall), sedangkan musim dingin memiliki penyewaan terendah.
    * Kemungkinan penyebab: Cuaca lebih nyaman di musim gugur, sementara di musim dingin cuaca terlalu ekstrem untuk bersepeda.

## 2. Pengaruh Waktu terhadap Penyewaan Sepeda

* Jumlah penyewaan tertinggi terjadi pada jam sibuk (pagi dan sore hari), sekitar 07:00 - 09:00 dan 17:00 - 19:00.
    * Kemungkinan penyebab: Banyak orang menggunakan sepeda sebagai alat transportasi untuk bekerja atau pulang kerja.

## 3. Kondisi Cuaca Berpengaruh Signifikan

* Cuaca cerah memiliki penyewaan terbanyak, sedangkan hujan dan salju mengurangi jumlah penyewa secara drastis.
    * Implikasi: Penyedia layanan sepeda bisa memberikan promo atau perlindungan ekstra saat cuaca buruk untuk meningkatkan penggunaan.

## 4. Pengaruh Faktor Lingkungan (Suhu, Kelembaban, dan Kecepatan Angin)

* Suhu sedang (antara dingin dan panas) memiliki penyewaan tertinggi, sedangkan suhu ekstrem mengurangi minat pengguna.
* Kelembaban yang sangat tinggi serta kecepatan angin yang tinggi menurunkan jumlah penyewa.
    * Kemungkinan penyebab: Cuaca nyaman (tidak terlalu panas atau dingin) lebih menarik bagi pengguna sepeda.

## 5. Tren Penyewaan Berdasarkan Bulan dan Hari

* Bulan dengan penyewaan tertinggi adalah bulan September dan Oktober, sedangkan bulan dengan penyewaan terendah adalah Januari dan Februari.
* Hari kerja memiliki lebih banyak penyewaan dibanding akhir pekan, menunjukkan bahwa banyak pengguna menggunakan sepeda untuk transportasi kerja.
    * Implikasi: Bisa dilakukan optimasi jumlah sepeda berdasarkan musim dan waktu tertentu untuk meningkatkan efisiensi.

## Analisis Lanjutan (Opsional)

### RFM Analysis:

```python
# Menghitung Recency: Seberapa baru pengguna menyewa sepeda (dari tanggal terakhir)
hour_df["dteday"] = pd.to_datetime(hour_df["dteday"])
latest_date = hour_df["dteday"].max()
rfm = hour_df.groupby("dteday").agg(
    Recency=("dteday", lambda x: (latest_date - x.max()).days),
    Frequency=("dteday", "count"),
    Monetary=("cnt", "sum")
).reset_index()

# Menampilkan hasil
print(rfm.head())
```

```python
# Pastikan kolom dteday dalam format datetime
hour_df["dteday"] = pd.to_datetime(hour_df["dteday"])

# Menghitung nilai maksimum tanggal dalam dataset (tanggal terbaru)
latest_date = hour_df["dteday"].max()

# Menghitung RFM
rfm = hour_df.groupby("dteday").agg(
    Recency=("dteday", lambda x: (latest_date - x.max()).days),
    Frequency=("dteday", "count"),
    Monetary=("cnt", "sum")
).reset_index()
# Set style seaborn terbaru
sns.set_theme(style="whitegrid", palette="muted")
```

```python
plt.figure(figsize=(12, 5))
sns.histplot(rfm["Monetary"], bins=20, kde=True, color="g")
plt.title("Distribusi Monetary (Histogram)", fontsize=14)
plt.xlabel("Total Penyewaan Sepeda", fontsize=12)
plt.ylabel("Frekuensi", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.scatterplot(x=rfm["Frequency"], y=rfm["Monetary"], color="purple")
plt.title("Hubungan Frequency vs. Monetary", fontsize=14)
plt.xlabel("Frequency (Jumlah Transaksi per Hari)", fontsize=12)
plt.ylabel("Monetary (Total Sepeda Disewa)", fontsize=12)
plt.show()
```

## Conclusion

## Distribusi Moneter (Histogram)
* Histogram menunjukkan distribusi bimodal dari total penyewaan sepeda, mengindikasikan adanya dua kelompok permintaan yang berbeda.
* Puncak pertama menunjukkan frekuensi tertinggi penyewaan sepeda berada di kisaran 4000-5000 sepeda.
* Puncak kedua menunjukkan adanya periode dengan jumlah penyewaan yang sangat tinggi, di kisaran 7000-8000 sepeda.
* Distribusi ini sedikit miring ke kanan, menunjukkan adanya beberapa hari dengan jumlah penyewaan yang sangat tinggi.
